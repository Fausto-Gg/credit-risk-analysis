```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv("../data/credit_risk_dataset.csv")

```

```python
df.shape
df.info()
```

```python
df.describe()
```

```python
df["default"].value_counts(normalize=True)
```

```python
sns.countplot(x="default", data=df)
plt.title("Default distribution")
plt.show()
```

```python
num_cols = ["age", "income", "credit_score", "loan_amount", "interest_rate"]

df[num_cols].hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
cat_cols = ["employment_status", "loan_term_months", "has_default_history"]

for col in cat_cols:
    sns.countplot(x=col, hue="default", data=df)
    plt.title(f"{col} vs Default")
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols + ["default"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
